```
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, mean, variance):
    return 1 / np.sqrt(2 * np.pi * variance) * np.exp(-(x - mean)**2 / (2 * variance))

# Input parameters
mean = float(input("Enter the mean: "))
variance = float(input("Enter the variance: "))
# num_points = int(input("Enter the number of points: "))
num_points = 50

# Generate random x values within a specified range
x_values = np.random.uniform(mean - 3 * np.sqrt(variance), mean + 3 * np.sqrt(variance), num_points)

# Evaluate the Gaussian function for each x value
y_values = gaussian(x_values, mean, variance)

# Plot the Gaussian distribution
x_range = np.linspace(mean - 3 * np.sqrt(variance), mean + 3 * np.sqrt(variance), 1000)
plt.plot(x_range, gaussian(x_range, mean, variance), label='Gaussian Distribution')
plt.scatter(x_values, y_values, color='red', label='Randomly chosen points')

plt.title('Gaussian Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

# Display the lists
print("X values:", x_values)
print("Corresponding Y values:", y_values)
x_values_model_1 = x_values
y_values_model_1 = y_values
```

```
import tensorflow as tf
import numpy as np
import random
import time

# Set a seed for reproducibility
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

# Define your training and testing data
X_training_model_1 = np.array([x_values_model_1[i] for i in random_indices_10])
X_training_model_2 = np.array([x_values_model_2[i] for i in random_indices_10])

X_testing_model_1 = np.array([x_values_model_1[i] for i in range(50) if i not in random_indices_10])
X_testing_model_2 = np.array([x_values_model_2[i] for i in range(50) if i not in random_indices_10])

Y_training_model_1 = np.array([y_values_model_1[i] for i in random_indices_10])
Y_training_model_2 = np.array([y_values_model_2[i] for i in random_indices_10])

Y_testing_model_1 = np.array([y_values_model_1[i] for i in range(50) if i not in random_indices_10])
Y_testing_model_2 = np.array([y_values_model_2[i] for i in range(50) if i not in random_indices_10])

# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2, input_shape=(2,), activation='sigmoid'),  # 2 inputs, 2 outputs
])

# Compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')

# Measure training time
start_time = time.time()

# Train the model
model.fit(np.column_stack((X_training_model_1, X_training_model_2)), np.column_stack((Y_training_model_1, Y_training_model_2)), epochs=10000)

training_time = time.time() - start_time

# Evaluate on the test set
predictions = model.predict(np.column_stack((X_testing_model_1, X_testing_model_2)))

# Calculate mean average precision (mAP) manually
mAP = np.mean(np.abs(Y_testing_model_1 - predictions[:, 0]) + np.abs(Y_testing_model_2 - predictions[:, 1]))

print("Training Time:", training_time, "seconds")
print("mAP:", mAP)
```
