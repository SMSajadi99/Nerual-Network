```python
import numpy as np
from matplotlib.image import imread
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

# Load and preprocess the image
image_raw = imread("/content/face.jpg")
image_sum = image_raw.sum(axis=2) / 3.0  # Convert to grayscale
image_bw = image_sum / image_sum.max()

# Flatten the image for the Autoencoder
input_img = image_bw.flatten().reshape(1, -1)

# Define the Autoencoder model
encoding_dim = 150  # Number of neurons in the encoding layer

input_layer = Input(shape=(input_img.shape[1],))
encoded = Dense(encoding_dim, activation='relu')(input_layer)
decoded = Dense(input_img.shape[1], activation='sigmoid')(encoded)

autoencoder = Model(input_layer, decoded)

autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Train the Autoencoder
autoencoder.fit(input_img, input_img, epochs=50, batch_size=1)

# Encode and decode the image
encoded_imgs = autoencoder.predict(input_img)
decoded_imgs = encoded_imgs.reshape(image_bw.shape)

# Display the original and reconstructed images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image_bw, cmap=plt.cm.gray)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(decoded_imgs, cmap=plt.cm.gray)
plt.title('Reconstructed Image')

plt.show()
```

```python
import numpy as np
from matplotlib.image import imread
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from sklearn.decomposition import PCA, IncrementalPCA

# Load and preprocess the image
image_raw = imread("/content/face.jpg")
image_sum = image_raw.sum(axis=2) / 3.0  # Convert to grayscale
image_bw = image_sum / image_sum.max()

# Flatten the image for the Autoencoder
input_img = image_bw.flatten().reshape(1, -1)

# Calculate the reduced encoding dimension for 95% variance
pca = PCA()
pca.fit(input_img)
var_cumu = np.cumsum(pca.explained_variance_ratio_)
k = np.argmax(var_cumu >= 0.95) + 1  # Add 1 to match 1-based indexing in PCA

# Define a deeper Autoencoder model
input_layer = Input(shape=(input_img.shape[1],))
encoded = Dense(256, activation='relu')(input_layer)
encoded = Dense(128, activation='relu')(encoded)
encoded = Dense(64, activation='relu')(encoded)

decoded = Dense(128, activation='relu')(encoded)
decoded = Dense(256, activation='relu')(decoded)
decoded = Dense(input_img.shape[1], activation='sigmoid')(decoded)

autoencoder = Model(input_layer, decoded)

autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Train the Autoencoder
autoencoder.fit(input_img, input_img, epochs=50, batch_size=1)

# Encode and decode the image
encoded_imgs = autoencoder.predict(input_img)
decoded_imgs = encoded_imgs.reshape(image_bw.shape)

# Display the original, encoded, and reconstructed images
plt.figure(figsize=(18, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image_bw, cmap=plt.cm.gray)
plt.title('Original Image')

# Encoded Image
plt.subplot(1, 3, 2)
plt.imshow(encoded_imgs.reshape(1, -1).reshape(image_bw.shape), cmap=plt.cm.gray)
plt.title('Compressed Image')

# Reconstructed Image
plt.subplot(1, 3, 3)
plt.imshow(decoded_imgs, cmap=plt.cm.gray)
plt.title('Reconstructed Image')

plt.show()

# Calculate Signal-to-Noise Ratio (SNR)
signal_power = np.sum(image_bw**2)
noise_power = np.sum((image_bw - decoded_imgs)**2)
snr = 10 * np.log10(signal_power / noise_power)

print("Signal-to-Noise Ratio (SNR): {:.2f} dB".format(snr))
```

```python
import numpy as np
from matplotlib.image import imread
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from sklearn.decomposition import PCA
from skimage.metrics import peak_signal_noise_ratio

# Load and preprocess the image
image_raw = imread("/content/face.jpg")
image_bw = image_raw.sum(axis=2) / 3.0
image_bw = image_sum / image_sum.max()

# Flatten the image for PCA
input_img_pca = image_bw.flatten().reshape(1, -1)

# Apply PCA for dimensionality reduction
n_components = 1
pca = PCA(n_components=n_components)
input_img_pca_reduced = pca.fit_transform(input_img_pca)

# Define an Autoencoder model
input_layer = Input(shape=(n_components,))
encoded = Dense(8, activation='relu')(input_layer)
decoded = Dense(n_components, activation='sigmoid')(encoded)

autoencoder = Model(input_layer, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Train the Autoencoder
autoencoder.fit(input_img_pca_reduced, input_img_pca_reduced, epochs=50, batch_size=1)

# Encode and decode the image
encoded_imgs = autoencoder.predict(input_img_pca_reduced)
decoded_imgs = pca.inverse_transform(encoded_imgs)

# Reshape decoded image to original shape
decoded_imgs = decoded_imgs.reshape(image_bw.shape)

# Display the original, encoded, and reconstructed images
plt.figure(figsize=(18, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image_bw, cmap=plt.cm.gray)
plt.title('Original Image')

# Encoded Image
plt.subplot(1, 3, 2)
compressed_img = np.resize(encoded_imgs.reshape(1, -1), image_bw.shape)
plt.imshow(compressed_img, cmap=plt.cm.gray)
plt.title('Compressed Image')

# Reconstructed Image
plt.subplot(1, 3, 3)
plt.imshow(decoded_imgs, cmap=plt.cm.gray)
plt.title('Reconstructed Image')

plt.show()

# Calculate Peak Signal-to-Noise Ratio (PSNR)
psnr = peak_signal_noise_ratio(image_bw, decoded_imgs)
print("Peak Signal-to-Noise Ratio (PSNR): {:.2f} dB".format(psnr))
```
